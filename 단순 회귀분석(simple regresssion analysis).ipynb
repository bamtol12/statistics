###단순 회귀분석(simple regression analysis)
:하나의 독립변수로 하나의 종속변수를 설명하는 모형.

독립변수는 $x_1, x_2, x_3, ... , x_n$ 로 나타내고
종속변수는 $y_1, y_2, y_3, ... , y_n$ 로 나타낸다.

통계적으로 해석 가능한지 알 수 있음.
두 변수 간의 인과관계를 조사하는 분석방법.

$y_i = \beta_0 + \beta_1x_i$

잔차(residual) e들의 '제곱의 합'이 최소가 되게 하는 직선이다.
(잔차들의 합은 항상 0)

$\beta_0, \beta_1$을 회귀계수라고 함.

###최소제곱법(least square estimation)
: 회귀계수의 추정

오차의 제곱을 더한 후 이의 최솟값을 찾는다. 

$
\displaystyle
Q = \sum_{i=1}^n(y_i - \beta_0 - \beta_1x_i)^2$

최솟값을 찾기 위해 위 식을 미분 후 연립.

$
\\
\\
$

-----

$
\\
\\
$

$
\displaystyle
{\Delta Q\over\Delta\beta_0} = -2\sum_{i=1}^n(y_i - \beta_0 - \beta_1x_i) = 0$

$
\displaystyle
\Rightarrow -2(\sum_{i=1}^n y_i - n\beta_0 - \beta_1\sum_{i=1}^n x_i) = 0$

$
\displaystyle
\Rightarrow \sum_{i=1}^n = n\beta_0 + \beta_1\sum_{i=1}^n x_i$

$
\\
$

$
\displaystyle
{\Delta Q\over\Delta\beta_1} = -2\sum_{i=1}^n x_i(y_i - \beta_0 - \beta_1x_i) = 0$

$
\displaystyle
\Rightarrow -2(\sum_{i=1}^n x_i y_i - \beta_0\sum_{i=1}^n x_i - \beta_1\sum_{i=1}^n x_i^2) = 0$

$
\displaystyle
\Rightarrow \sum_{i=1}^n x_i y_i = \beta_0\sum_{i=1}^n x_i + \beta_1\sum_{i=1}^n x_i^2)$

$
\\
$

###정규 방정식(normal equation):

$
\displaystyle
\sum_{i=1}^n = n\beta_0 + \beta_1\sum_{i=1}^n x_i$
$
\displaystyle
\sum_{i=1}^n x_i y_i = \beta_0\sum_{i=1}^n x_i + \beta_1\sum_{i=1}^n x_i^2$

윗식에 $\displaystyle \sum_{i=1}^n x_i$, 아랫식에 $n$을 곱하여 연립한다. 

$
\\
\\
$

$

= {Sxy\over Sxx}
